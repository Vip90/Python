```python
import tkinter as tk #importing external library for Userinterface
import loaddictusersbooks as load #importing loading module for user_preference dictionary 

root= tk.Tk()# main window

canvas1 = tk.Canvas(root, width=1000, height=500)#setting size for main window
canvas1.pack()

entry1 = tk.Entry(root) #textbox for user1
entry2 = tk.Entry(root) #textbox for user 2

widget = tk.Label(canvas1, text='Enter first User/Book ID here:', fg='black')#label for better understanding 
widget.pack()

widget1 = tk.Label(canvas1, text='Enter Second User/Book ID here:', fg='black')#label for better understanding 
widget1.pack()


canvas1.create_window(80, 140, window=widget)#setting position of label and textbox
canvas1.create_window(300, 140, window=entry1)#setting position of label and textbox
canvas1.create_window(650, 140, window=entry2)#setting position of label and textbox
canvas1.create_window(470, 140, window=widget1)#setting position of label and textbox


#below function is ued to close the main application     
def quit():
    root.destroy()

    
    
# N similar Users and book calculated here   
    
    
    
def openNewWindow():
    
    #below function is used to close the new window
    def quitnsimilarwindow():
        newWindow.destroy()
    
    #below function us used to compare n similar books
    def similarbooks():
        try:
            Bookdict=load.load_dataset_module_bookdetails()
            listbooktitle=[]#creates an empty list to store booktitle to compare later
            listbookwriter=[]#creates an empty list to store writer to compare later
            Bookisbn=entrynbook.get()#getting the bookisbn from textbox
            for title in Bookdict[Bookisbn]:
                listbooktitle.append(title)
                listbookwriter.append(Bookdict[Bookisbn][title])
            for isbn,title in Bookdict.items():
                for title,writer in title.items():
                    if writer in listbookwriter and isbn!=Bookisbn:
                        labeloutput = tk.Label(newWindow, text='BookID:'+' '+' '+isbn+''+' '+'With Title'+' '+title+' '+''+'has been written by the same author'+' '+writer)
                        labeloutput.pack()
                    
        except KeyError:
            labeloutput = tk.Label(newWindow, text='No Book Id found or wrong or no input detected')
            labeloutput.pack()
        
#below function is used to compare n similar users
    def nsimilarusers():
        try:
            user_pref=load.load_dataset_module_userrating()
            listbookisbn=[]#empty list to store bookisbn to be searched later
            listbookrating=[]#empty list to store bookrating to be searched later
            userid=entrynuser.get()#getting the userid from textbox
            for bookisbn in user_pref[userid]:
                listbookisbn.append(bookisbn)
                listbookrating.append(user_pref[userid][bookisbn])
            for useridkey,bookvalues in user_pref.items():
                for bookisbn,rating in bookvalues.items():
                    #below statement is used to compare similar rating and bookisbn and making sure the userid is not the same as the searched one
                    if bookisbn in listbookisbn and rating in listbookrating and useridkey!=userid:
                        #labelvariable.set('Userid:'+' '+' '+useridkey+''+' '+'has rated books with id'+' '+bookisbn+' '+''+'with the same ratings'+' '+rating)
                        #labeloutput1=tk.Label(newWindow,textvariable=labelvariable)
                        labeloutput = tk.Label(newWindow, text='Userid:'+' '+' '+useridkey+''+' '+'has rated books with id'+' '+bookisbn+' '+''+'with the same ratings'+' '+rating)
                        labeloutput.pack()
                
                        
                        
        except KeyError:
            labeloutput = tk.Label(newWindow, text='No User Id found or wrong or no input detected')
            labeloutput.pack()
    
    #below code is used for designing purpose only for the new window
    newWindow = tk.Toplevel(canvas1)
 
    
    newWindow.title("New Window")
 
    
    newWindow.geometry("200x200")
 
    
    tk.Label(newWindow,
          text ="This is a new window").pack()
    widgetn = tk.Label(newWindow, text='Enter the UserID to n find similar users:', fg='black')
    widgetn.pack()
    
    entrynuser = tk.Entry(newWindow) 
    entrynuser.pack()
    
    button7 = tk.Button(newWindow,text='Calculate n Similar Users', command=nsimilarusers)
    button7.pack()
    
    widgetnbook = tk.Label(newWindow, text='Enter the BookISBN to n find similar users:', fg='black')
    widgetnbook.pack()
    
    entrynbook = tk.Entry(newWindow)
    entrynbook.pack()
    button8 = tk.Button(newWindow,text='Calculate n Similar Books', command=similarbooks)
    button8.pack()
    button9 = tk.Button(newWindow,text='Back to Main Window', command=quitnsimilarwindow)
    button9.pack()
 
    
#Similarity between users calculated here:
    
def jaccardsimilarity():
    try:
        labelvariable=tk.StringVar()#variable declare to change the label text 
        user_pref=load.load_dataset_module_userrating()#loading the userpreference dictionary
        rater1 = entry1.get()#getting output from textbox for user1
        rater2=entry2.get()#getting output from textbox for user2
        userpref1=user_pref[rater1]#store the bookisbn and rating for the first user
        userpref2=user_pref[rater2]#store the bookisbn and rating  for the second user
        set1=set(userpref1)#creates a set and stores the content of previously created list for comparison
        set2=set(userpref2)#creates a set and stores the content of previously created list for comparison
        n1=len(set1.intersection(set2))#find intersection between user1 and user2
        n2=len(set1.union(set2))#find union between user1 and user2
       
        #labeloutput = tk.Label(root, text=n1/n2)
        #labeloutput=tk.Label(root, text=n1/n2)
        similarityscore=n1/n2
        labelvariable.set('Similarity Score between the mentioned users by jaccards method is '+' '+str(round(similarityscore,3)))
        labeloutput=tk.Label(root,textvariable=labelvariable)
        #canvas1.create_window(500, 100, window=labeloutput)
        labeloutput.pack()
    except KeyError as k:
        labelvariable.set('No User Id found or wrong or no input detected')
        labeloutput=tk.Label(root,textvariable=labelvariable)
        #labeloutput = tk.Label(root, text='No User Id found or wrong or no input detected')
        #canvas1.create_window(500, 100, window=labeloutput)
        labeloutput.pack()
    except ZeroDivisionError:
        labelvariable.set('No Similarity found')
        labeloutput=tk.Label(root,textvariable=labelvariable)
        #labeloutput = tk.Label(root, text='No Similarity found')
        #canvas1.create_window(500, 100, window=labeloutput)
        labeloutput.pack()
    except ValueError:
        labelvariable.set('Wrong input detected for numbers')
        labeloutput=tk.Label(root,textvariable=labelvariable)
        #labeloutput = tk.Label(root, text='Wrong input detected for numbers')
        canvas1.create_window(500, 100, window=labeloutput)
        labeloutput.pack()
    except KeyboardInterrupt:
        labelvariable.set('Application has been closed by User Intervention')
        labeloutput=tk.Label(root,textvariable=labelvariable)
        #labeloutput = tk.Label(root, text='Application has been closed by User Intervention')
        #canvas1.create_window(500, 100, window=labeloutput)
        labeloutput.pack()
def eculidean():
    try:
        labelvariable=tk.StringVar()#variable declare to change the label text 
        user_pref=load.load_dataset_module_userrating()#for loading the user preference dictionary
        rater1 = entry1.get()#getting output from textbook for user 1
        rater2=entry2.get()#getting output from textbook for user 2
        total=0
        from math import sqrt
        for book in user_pref[rater1]:
            if book in user_pref[rater2]:
                total=total+pow(int(user_pref[rater1][book])-int(user_pref[rater2][book]),2)
        similaityscore=1/sqrt(total)
        labelvariable.set('Similarity Score between the mentioned users by eculidean method is  '+' '+str(round(similaityscore,3)))
        labeloutput=tk.Label(root,textvariable=labelvariable)
        #labeloutput = tk.Label(root, text=1/sqrt(total))
        #canvas1.create_window(500, 100, window=labeloutput)
        labeloutput.pack()
    except KeyError:
        labelvariable.set('No User Id found or wrong or no input detected')
        labeloutput=tk.Label(root,textvariable=labelvariable)
        #labeloutput = tk.Label(root, text='No User Id found or wrong or no input detected')
        #canvas1.create_window(500, 100, window=labeloutput)
        labeloutput.pack()
    except ZeroDivisionError:
        labelvariable.set('No Similarity found')
        labeloutput=tk.Label(root,textvariable=labelvariable)
        #labeloutput = tk.Label(root, text='No Similarity found')
        #canvas1.create_window(500, 100, window=labeloutput)
        labeloutput.pack()
    except ValueError:
       
        labelvariable.set('Wrong input detected for numbers')
        labeloutput=tk.Label(root,textvariable=labelvariable)
        #labeloutput = tk.Label(root, text='Wrong input detected for numbers')
        #canvas1.create_window(500, 100, window=labeloutput)
        labeloutput.pack()
    except KeyboardInterrupt:
        labelvariable.set('Application has been closed by User Intervention')
        labeloutput=tk.Label(root,textvariable=labelvariable)
        #labeloutput = tk.Label(root, text='Application has been closed by User Intervention')
        #canvas1.create_window(500, 100, window=labeloutput)
        labeloutput.pack()
def manhattan():
    try:
        labelvariable=tk.StringVar()
        user_pref=load.load_dataset_module_userrating()
        rater1 = entry1.get()
        rater2=entry2.get()
        a=[]
        b=[]
        #print(user_perf[rater1])
        for book in user_pref[rater1]:
            #print(book)
            if book in user_pref[rater2]:
                x=int(user_pref[rater1][book])
                y=int(user_pref[rater2][book])
                a.append(x)
                b.append(y)
        similarityscore=1/sum(abs(val1-val2) for val1, val2 in zip(a,b))
        #labeloutput = tk.Label(root, text=1/sum(abs(val1-val2) for val1, val2 in zip(a,b)))
        labelvariable.set('Similarity Score between the mentioned users by manhattan method is  '+' '+
                          str(round(similarityscore,3)))
        labeloutput=tk.Label(root,textvariable=labelvariable)
        #canvas1.create_window(500, 280, window=labeloutput)
        labeloutput.pack()
    except KeyError:
        labelvariable.set('No User Id found or wrong or no input detected')
        labeloutput=tk.Label(root,textvariable=labelvariable)
        #labeloutput = tk.Label(root, text='No User Id found or wrong or no input detected')
        #canvas1.create_window(500, 230, window=labeloutput)
        labeloutput.pack()
    except ZeroDivisionError:
        labeloutput = tk.Label(root, text="")
        labeloutput = tk.Label(root, text='No Similarity found')
        #canvas1.create_window(500, 100, window=labeloutput)
        labeloutput.pack()
    except ValueError:
        labeloutput = tk.Label(root, text="")
        labeloutput = tk.Label(root, text='Wrong input detected for numbers')
        canvas1.create_window(500, 100, window=labeloutput)
        labeloutput.pack()
    except KeyboardInterrupt:
        labeloutput = tk.Label(root, text="")
        labeloutput = tk.Label(root, text='Application has been closed by User Intervention')
        #canvas1.create_window(500, 100, window=labeloutput)
        labeloutput.pack()
        
from math import sqrt

def avg_distance():
    try:
        labelvariable=tk.StringVar()
        user_pref=load.load_dataset_module_userrating()
        rater1 = entry1.get()
        rater2 = entry2.get()
        a=[]
        b=[]
        for book in user_pref[rater1]:
            if book in user_pref[rater2]:
                x=int(user_pref[rater1][book])
                y=int(user_pref[rater2][book])
                a.append(x)
                b.append(y)
        #print(a)
        #print(b)
        n = len(a)
        #print(n)
        dist = 0
        for i in range(n):
            ai = a[i]
            bi = b[i]
            for j in range(i+1,n):
                da = a[j]-ai
                db = b[j]-bi
                dist += sqrt(da*da+db*db)
        #return 1/2.0*dist/(n*(n-1))
        similarityscore=1/(2.0*dist/(n*(n-1)))
        labelvariable.set('Similarity Score between the mentioned users by Average Distance method is  '+' '
                          +str(round(similarityscore,3)))
        labeloutput=tk.Label(root,textvariable=labelvariable)
        #labeloutput = tk.Label(root, text='Similarity Score between the mentioned users by Average Distance method is  '+' '+str(1/2.0*dist/(n*(n-1))))
        #canvas1.create_window(500, 230, window=labeloutput)
        labeloutput.pack()
    
    except KeyError:
        labelvariable.set('No User Id found or wrong or no input detected')
        #labeloutput = tk.Label(root, text='No User Id found or wrong or no input detected')
        #canvas1.create_window(500, 100, window=labeloutput)
        labeloutput=tk.Label(root,textvariable=labelvariable)
        labeloutput.pack()
    except ZeroDivisionError:
        labelvariable.set('No Similarity found')
        #labeloutput = tk.Label(root, text='No Similarity found')
        labeloutput=tk.Label(root,textvariable=labelvariable)
        #canvas1.create_window(500, 100, window=labeloutput)
        labeloutput.pack()
    except ValueError:
        labelvariable.set('Wrong input detected for numbers')
        labeloutput = tk.Label(root, text='Wrong input detected for numbers')
        labeloutput=tk.Label(root,textvariable=labelvariable)
        #canvas1.create_window(500, 100, window=labeloutput)
        labeloutput.pack()
    except KeyboardInterrupt:
        labelvariable.set('Application has been closed by User Intervention')   
        #labeloutput = tk.Label(root, text='Application has been closed by User Intervention')
        labeloutput=tk.Label(root,textvariable=labelvariable)
        #canvas1.create_window(500, 100, window=labeloutput)
        labeloutput.pack()
        
def cosine_similarity():
    try:
        labelvariable=tk.StringVar()#variable declared to change the text in label
        user_pref=load.load_dataset_module_userrating()#loading the user_preference dictionary
        rater1 = entry1.get()#getting output from user 1
        rater2 = entry2.get()#getting output from user 2
        a=[]#creating an empty list for user1 data
        b=[]#creating an empty list for user2 data
        for book in user_pref[rater1]:
            if book in user_pref[rater2]:
                e=int(user_pref[rater1][book])#converting to integer values for calculation
                f=int(user_pref[rater2][book])
                a.append(e)
                b.append(f)
        similarityscore=1/(sum([i*j for i,j in zip(a, b)])/(sqrt(sum([i*i for i in a]))
                                                                 * sqrt(sum([i*i for i in b]))))
        #return sum([i*j for i,j in zip(a, b)])/(sqrt(sum([i*i for i in a]))* sqrt(sum([i*i for i in b])))
        labelvariable.set('Similarity Score between the mentioned users by Cosine method is  '+' '+
                          str(round(similarityscore,3)))
        labeloutput=tk.Label(root,textvariable=labelvariable)
        #labeloutput = tk.Label(root, text='Similarity Score between the mentioned users by Cosine Distance method is  '+' '+str(sum([i*j for i,j in zip(a, b)])/(sqrt(sum([i*i for i in a]))* sqrt(sum([i*i for i in b])))))
        labeloutput.pack()
    except KeyError:
        labelvariable.set('No User Id found or wrong or no input detected')
        labeloutput=tk.Label(root,textvariable=labelvariable)
        #labeloutput = tk.Label(root, text='No User Id found or wrong or no input detected')
        #canvas1.create_window(500, 100, window=labeloutput)
        labeloutput.pack()
    except ZeroDivisionError:
        labelvariable.set('No Similarity found')
        labeloutput=tk.Label(root,textvariable=labelvariable)
        #labeloutput = tk.Label(root, text='No Similarity found')
        #canvas1.create_window(500, 100, window=labeloutput)
        labeloutput.pack()
    except ValueError:
        labelvariable.set('Wrong input detected for numbers')
        labeloutput=tk.Label(root,textvariable=labelvariable)
        #labeloutput = tk.Label(root, text='Wrong input detected for numbers')
        #canvas1.create_window(500, 100, window=labeloutput)
        labeloutput.pack()
    except KeyboardInterrupt:
        labelvariable.set('Application has been closed by User Intervention')
        labeloutput=tk.Label(root,textvariable=labelvariable) 
        #labeloutput = tk.Label(root, text='Application has been closed by User Intervention')
        #canvas1.create_window(500, 100, window=labeloutput)
        labeloutput.pack()
        
        

        
        
#below code is used for designing purpose only creating button positioning them and calling various functions assocaited.       
    
button1 = tk.Button(text='Calulcate Similarity via Jaccards Method for Users', command=jaccardsimilarity)
button2 = tk.Button(text='Calulcate Similarity via eculidean Method for Users', command=eculidean)
button3 = tk.Button(text='Calulcate Similarity via manhattan Method for Users', command=manhattan)
button4 = tk.Button(text='Calulcate Similarity via Average Distance Method for Users', command=avg_distance)
button5 = tk.Button(text='Calulcate Similarity via Cosine Method for Users', command=cosine_similarity)
button6 = tk.Button(text='Calculate n Similar Users/Books for Users', command=openNewWindow)


button7 = tk.Button(text='Quit', command=quit)


canvas1.create_window(370, 180, window=button1)#set position for button1
canvas1.create_window(370, 220, window=button2)#set position for button2
canvas1.create_window(370, 260, window=button3)#set position for button3
canvas1.create_window(370, 300, window=button4)#set position for button4
canvas1.create_window(370, 340, window=button5)#set position for button5
canvas1.create_window(370, 380, window=button6)#set position for button6
canvas1.create_window(370, 460, window=button7)#set position for button7







root.mainloop()
```
